```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,1,10)
m=0.6
c=25
y=m*x+c
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,1,10)
y=np.sin(2*x)

fig, a=plt.subplots()

a.plot( x, y, linewidth=3.0)

a.set(xlim=(0,8),xticks=np.arange(1,8)
,ylim=(0,8),yticks=np.arange(1,8))

plt.show()
```

# **Thermistor characteristics**

```python
import matplotlib.pyplot as plt
import numpy as np

temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120,130])
res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608,420])

plt.scatter(temp,res)
plt.show()
```

# **Garnishing Graphs**

```python
import matplotlib.pyplot as plt
import numpy as np

temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120,130])
res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608,420])

plt.plot(temp,res)
plt.title=("Thermister Characterstics")
plt.xlabel("temp")
plt.ylabel("res")
plt.grid()
plt.show()
```

# **Bar and Pie Chart**

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(["karnataka","AP","TN","TG"])
y=np.array([65,80,89,45])

plt.bar(x,y)
plt.show()

plt.pie(y)
plt.show()
```
