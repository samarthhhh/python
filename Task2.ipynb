# **Question.A**

```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
df = pd.read_excel('/content/new1.xlsx')
df = df.values

x1 = df[:, 0]
x2= df[:, 1]
x3= df[:, 2]
x4 = df[:, 3]
y = df[:, 4]
print("y :",y)

print("x1 :" ,x1)
print("x2:",x2)
print("x3:",x3)
print("x4:",x4)

x1=x1.reshape(-1,1)
x2=x2.reshape(-1,1)
x3=x3.reshape(-1,1)
x4=x4.reshape(-1,1)
plt.scatter(x1,y)
plt.scatter(x2,y)
plt.scatter(x3,y)
plt.scatter(x4,y)
plt.xlabel("IV")
plt.ylabel("DV")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
slr1=LinearRegression()
slr1.fit(x1,y)
slr2=LinearRegression()
slr2.fit(x2,y)
slr3=LinearRegression()
slr3.fit(x3,y)
slr4=LinearRegression()
slr4.fit(x4,y)
pred1=slr1.predict(x1)
pred2=slr2.predict(x1)
pred3=slr3.predict(x3)
pred4=slr4.predict(x4)
from sklearn.metrics import mean_squared_error
mse1=mean_squared_error(y,pred1)
mse2=mean_squared_error(y,pred2)
mse3=mean_squared_error(y,pred3)
mse4=mean_squared_error(y,pred4)
print("mse: ",mse1,mse2,mse3,mse4)
print("INTERCEPT :" ,slr1.intercept_,slr2.intercept_,slr3.intercept_,slr4.intercept_)
print("slpoe :",slr1.coef_,slr2.coef_,slr3.coef_,slr4.coef_)
plt.plot(x1,pred1,color='red',marker='*')
plt.plot(x2,pred2,color='black',marker='.')
plt.plot(x3,pred3,color='blue',marker='+')
plt.plot(x4,pred4,color='violet',marker='*')

```
