# **Tuple Continuation**

```python
tuple1=("hubli","belgaum","hubli")
print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
print(tuple1[-1])
tuple6=tuple1+tuple1
print(tuple6)
tuple7=tuple6*2
print(tuple7)
y=tuple7.index('hubli')
print(y)
x=tuple7.count('hubli')
print(x)
```

# **Type casting**

```python
x=("hubli","belgaum","blore")
print(type(x))
y=list(x)
y[2]="mumbai"
x=tuple(y)
print(x)
```

# **Sets**

```python
set1={"red","blue","black","red"}
print(set1)
set1.add('orange')
print(set1)
set2={"ball","bat","wicket"}
set1.update(set2)
print(set1)
```
